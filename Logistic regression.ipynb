### Synthetic Data Generation (0.25 point)

```python
## We will use the sklearn datasets to experiment with Logistic regression.
## Read the documentation of the function make_classification at
## https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html
##
##
## We will also use the train_test_split function to split the dataset into train and test.
## Read the documentation of the function train_test_split at
## https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html


from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

## create the following variables to generate the dataset
## refer the decumentaion to understand what each variable does

## n_samples: It is an Integer,total number of samples.
##n_features: Integer, total number of features
##n_informative: Integer, number of informative features.number of gaussian cluster sloctaed around the vertices of a hypercube in a subspace of dimension n_informative
##n_redundant :Integer, these  are generated as random linear combinations of the informative features.
##n_clusters_per_class: Integer,number of clusters per class.
##flip_y: Float,fraction of samples whose class is assigned randomly


## create a variable named n_samples
n_samples = 500
## create a variable named n_features
n_features = 2
## create a variable named n_informative
n_informative = 2
## create a variable named n_redundant
n_redundant = 0
## create a variable named n_clusters_per_class
n_clusters_per_class = 2
## create a variable named flip_y
flip_y = 0.1

## Generate synthetic data with errors (misclassified labels)
## Introduce 10% misclassified labels
## <write your code here>
x_class,y_class =  make_classification(n_samples = n_samples,n_features =n_features ,n_informative = n_informative,n_redundant = n_redundant,n_clusters_per_class = n_clusters_per_class,flip_y = flip_y)
print(x_class,y_class)
```

# Logistic regression (1.5 point)

```python
# Sigmoid Activation Function
import numpy as np
def sigmund(z):
    sig = 1 / (1 + np.exp(-z))
    return sig
```

```python
## Implement the loss function
def cross_entropy_loss(y_true, y_pred):
   epsilon = 1e-15
   y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
   loss =  -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
   return loss
```

```python
## Implement a function for prediction
def predict(X, weights):
    z = np.dot(X, weights)
    y_pred = sigmund(z)
    return [1 if p >= 0.5 else 0 for p in y_pred]
```

```python
## Implement gradient descent for logistic regression from scratch
def gradient_descent(X, y, learning_rate, num_epochs):

    m, n = X.shape
    weights = np.zeros(n)
    bias = []

    for epoch in range(num_epochs):
        z = np.dot(X, weights)
        y_pred = sigmund(z)
        error = y_pred - y
        gradient = np.dot(X.T, error) / m
        weights -= learning_rate * gradient
        loss = cross_entropy_loss(y, y_pred)
        bias.append(loss)
    return weights, bias
```

```python
## Train and test the model
#Train
learning_rate = 0.01
epochs = 1000

weights, bias = gradient_descent(x_class_train, y_class_train, learning_rate,epochs)
#Test

y_pred_scratch = predict(x_class_test, weights)
print(y_pred_scratch)
```

```python
## Plot loss vs epoch
plt.plot(range(epochs), bias)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Function vs. Epoch (Training Data)')
plt.show()
```
